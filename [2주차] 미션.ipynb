## 📌Q4. 평균 기대수명이 점점 길어지면서 실버 의료 산업도 주목받고 있습니다. 여러분은 실버 의료 산업과 관련된 스타트업에 취업했습니다. 지도를 시각화하여 '노인/치매병원'이 주로 어디에 위치하고 있는지를 찾아보고자 합니다.

- folium 을 통해 지도에 전국의 '노인/치매병원'을 표시해 주세요!
- 다음과 같이 시각화 하며, folium 의 문서를 참고하여 다른 기능을 사용하여 좀 더 멋지게 지도를 꾸며도 좋습니다.
- folium 문서 : https://python-visualization.github.io/folium/quickstart.html

```python
# '상권업종소분류명'이 '노인|치매'인 데이터를 가져와 senior에 저장
senior=df[df['상권업종소분류명'].str.contains('노인|치매')]
senior
```

```python
import folium

# senoir 데이터의 위도, 경도 평균값을 중심으로 지도 생성
map=folium.Map(location=[senior['위도'].mean(),senior['경도'].mean()], zoom_start=7)

# 지도에 '상호명'과 '도로명주소'를 포함한 마커 추가하기
for n in senior.index:
    name=senior.loc[n,'상호명'] # n번 행의 상호명
    address=senior.loc[n,'도로명주소'] # n번 행의 도로명주소
    popup=f"{name}-{address}" # 상호명과 도로명주소 이어붙이기
    location=[senior.loc[n,'위도'],senior.loc[n,'경도']] # n번 행의 위도, 경도
    folium.Marker(
        location=location, # 위도 경도 위치에
        popup=popup, # 상호명과 도로명 주소 popup 띄우기
    ).add_to(map) # 마커를 지도에 추가하기

map
```
