```python
df_raw.melt(id_vars=['국가(대륙)별','상품군별','판매유형별'], var_name='기간',value_name='백만원')
```

```python
df=df_raw.melt(id_vars=['국가(대륙)별','상품군별','판매유형별'], var_name='기간',value_name='백만원')
df.shape
```

```python
df.info()
```

## 데이터 전처리

### 기간에서 연도를 분리하기

```python
# '기간' 컬럼 확인 -> '.'으로 '연도'와 '분기' 구분 & object 형식
df['기간']
```

```python
# '2019.4/4'를 '.'을 기준으로 분리
'2019.4/4'.split('.')
```

```python
# '2019.4/4'를 '.'을 기준으로 분리한 첫번째 요소
'2019.4/4'.split('.')[0]
```

```python
# '2019.4/4'를 '.'을 기준으로 분리한 첫번째 요소의 데이터 타입 확인 -> string
type('2019.4/4'.split('.')[0])
```

```python
# 정수형으로 바꿔주기
int('2019.4/4'.split('.')[0])
```

```python
# 위와 같은 과정으로 '기간' 컬럼에서 '연도'만 분리하기
df['연도']=df['기간'].map(lambda x : int(x.split('.')[0]))
df['연도']
```

```python
df.head()
```

```python
# '2019.1/4'를 '.'을 기준으로 분리한 두번째 요소
'2019.1/4'.split('.')[1]
```

```python
# '2019.1/4'를 '.'을 기준으로 분리한 두번째 요소를 '/'를 기준으로 다시 분리
'2019.1/4'.split('.')[1].split('/')
```

```python
# '2019.1/4'를 '.'을 기준으로 분리한 두번째 요소를 '/'를 기준으로 다시 분리한 첫번째 요소가 '분기'
'2019.1/4'.split('.')[1].split('/')[0]
```

```python
# 데이터 형식 확인 -> string
type('2019.1/4'.split('.')[1].split('/')[0])
```

```python
# 정수형으로 바꿔주기
int('2019.1/4'.split('.')[1].split('/')[0])
```

```python
# 위와 같은 과정으로 '기간'에서 '분기'만 분리하기
df['분기']=df['기간'].map(lambda x : int(x.split('.')[1].split('/')[0]))
df['분기']
```

```python
df.head()
```

### 금액을 수치데이터로 표현하기 위해 데이터 타입 변경하기

```python
# '백만원' 컬럼 확인
df['백만원']
```

```python
# '-'문자를 결측치로 변경
df['백만원'].replace('-',pd.np.nan)
```

```python
# 결측치는 float 타입이므로 데이터형식을 바꿔주고 '백만원' 컬럼에 반영
df['백만원']=df['백만원'].replace('-',pd.np.nan).astype(float)
df['백만원']
```

```python
df.head()
```

### 필요없는 데이터 제거하기

```python
# 합계 데이터는 따로 구할 수 있으므로 합계 데이터를 제외한 데이터만 사용
df=df[(df['국가(대륙)별']!='합계') & (df['상품군별']!='합계')].copy()
df
```

```python
# 결측치 확인
df.isnull().sum()
```

## K-Beauty 시각화

### 전체 상품군 판매액

```python
# '판매유형별' 데이터는 일부 기간에는 '계'만 존재하기 때문에 
# '판매유형별'=='계' 데이터만 가져와야 평균 값을 구하는 그래프에서 올바른 값 표현 가능
df_total=df[df['판매유형별']=='계'].copy()
df_total
```

```python
# 연도, 백만원 lineplot으로 그리기
sns.lineplot(data=df_total,x='연도',y='백만원')
```

```python
# 연도, 백만원 lineplot으로 그리고 상품군별로 다른 색상으로 표시
sns.lineplot(data=df_total,x='연도',y='백만원',hue='상품군별')
```

```python
# 연도, 백만원 lineplot으로 그리고 상품군별로 다른 색상으로 표시
# 그래프 밖에 legend 표시하기
sns.lineplot(data=df_total,x='연도',y='백만원',hue='상품군별')
plt.legend(bbox_to_anchor=(1.05,1),loc=2,borderaxespad=0.)
```

```python
# 위의 그래프를 자세하게 보기 위해 서브플롯으로 그리기
# relplot
# col_wrap: 한 줄에 몇개의 그래프를 나타낼 지 숫자 지정
sns.relplot(data=df_total,x='연도',y='백만원',hue='상품군별',
            kind='line',col='상품군별',col_wrap=4)
```

```python
# isin을 사용해 '화장품'만 제외하고 df_sub 에 저장
df_sub=df_total[~df_total['상품군별'].isin(['화장품'])].copy()
```

```python
# df_sub를 relplot을 활용해 서브플롯으로 그리기
sns.relplot(data=df_sub,x='연도',y='백만원',hue='상품군별',
            kind='line',col='상품군별',col_wrap=4)
```

```python
# '화장품'과 '의류 및 패션 관련상품'을 제외하고 다시 시각화하기
df_sub2=df_total[~df_total['상품군별'].isin(['화장품','의류 및 패션 관련상품'])].copy()
sns.relplot(data=df_sub2,x='연도',y='백만원',hue='상품군별',
            kind='line',col='상품군별',col_wrap=4)
```

### 화장품의 온라인쇼핑 해외직접판매액

```python
# '상품군별'이 '화장품'인 데이터를 df_cosmetic으로 저장
df_cosmetic=df_total[df_total['상품군별']=='화장품'].copy()
df_cosmetic
```

```python
df_cosmetic['상품군별'].unique()
```

```python
# '연도'와 '판매액'을 lineplot으로 그리고 '분기'별로 다른 색상으로 표현하기
plt.figure(figsize=(15,4)) # 그래프 크기
sns.lineplot(data=df_cosmetic,x='연도',y='백만원',hue='분기')
```

```python
# 화장품 '판매액'에 대한 '기간'별 금액 데이터 시각화
plt.figure(figsize=(15,4))
plt.xticks(rotation=30) # 글씨 기울기
sns.lineplot(data=df_cosmetic,x='기간',y='백만원')
```

```python
# 화장품 '판매액'에 대한 '기간'별 금액 데이터 시각화하고 '국가(대륙)별' 다른 색상으로 표현
plt.figure(figsize=(15,4))
plt.xticks(rotation=30) # 글씨 기울기
sns.lineplot(data=df_cosmetic,x='기간',y='백만원',hue='국가(대륙)별')
```

```python
# 중국을 제외한 '국가(대륙)별' 다른 색상으로 표현
plt.figure(figsize=(15,4))
plt.xticks(rotation=30) # 글씨 기울기
sns.lineplot(data=df_cosmetic[df_cosmetic['국가(대륙)별']!='중국'],
             x='기간',y='백만원',hue='국가(대륙)별')
```

```python
# 화장품 '판매액'에 대한 '기간'별 금액 데이터 시각화하고 '판매유형별' 다른 색상으로 표현
df_sub=df[df['판매유형별']!='계'].copy()

plt.figure(figsize=(15,4))
plt.xticks(rotation=30)
sns.lineplot(data=df_sub,x='기간',y='백만원',hue='판매유형별')
```

```python
# '면세점 이외'에 대한 시각화
df_sub=df[(df['판매유형별']!='계') & (df['판매유형별']!='면세점')].copy()


plt.figure(figsize=(15,4))
plt.xticks(rotation=30)
sns.lineplot(data=df_sub,x='기간',y='백만원',hue='판매유형별',ci=None)
```

### 의류 및 패션 관련상품 온라인쇼핑 해외직접판매액

```python
# '의류 및 패션 관련상품' 데이터의 '판매유형별'=='계'만 df_fashion으로 가져오기
df_fashion=df[(df['상품군별']=='의류 및 패션 관련상품') & (df['판매유형별']=='계')].copy()
df_fashion
```

```python
# '의류 및 패션 관련상품' 판매액에 대한 기간별 금액 데이터 시각화
# '국가(대륙)별' 다른 색상으로 표현
plt.figure(figsize=(15,4))
plt.xticks(rotation=30)
sns.lineplot(data=df_fashion,x='기간',y='백만원',hue='국가(대륙)별')
```

```python
# '의류 및 패션 관련상품' 판매액에 대한 기간별 금액 데이터 시각화
# '판매유형별' 다른 색상으로 표현
df_fashion2=df[(df['상품군별']=='의류 및 패션 관련상품') & (df['판매유형별']!='계')].copy()

plt.figure(figsize=(15,4))
plt.xticks(rotation=30)
sns.lineplot(data=df_fashion2,x='기간',y='백만원',hue='판매유형별',ci=None)
```

### 데이터 집계하기

```python
# pivot_table로 '국가(대륙)별', '연도'별 합계 금액을 표 형태로 구하기
df_fashion.pivot_table(index='국가(대륙)별',columns='연도',values='백만원',aggfunc='sum')
```

### 연산결과를 시각적으로 보기

```python
result=df_fashion.pivot_table(index='국가(대륙)별',columns='연도',values='백만원',aggfunc='sum')
```

```python
# pivot_table로 구한 결과를 값의 많고 적음에 따라 시각적으로 표현하기
plt.figure(figsize=(10,6))
sns.heatmap(result, # 데이터
            cmap='Blues', # 색상
            annot=True, # 값 표현
            fmt='.0f') # 소숫점 아래
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(result, # 데이터
            cmap='Blues_r', # 색상 -> # Blue_r'을 하면 값이 작을수록 진한 색

            annot=True, # 값 표현
            fmt='.0f') # 소숫점 아래
```
